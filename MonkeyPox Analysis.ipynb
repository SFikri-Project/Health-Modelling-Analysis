```python
import random

# Fungsi untuk menghasilkan data random di antara rentang 2,9 - 4
def generate_random_data():
    return random.uniform(2.9, 4)

# Jumlah data yang diinginkan
jumlah_data = 23

# Target rata-rata
rata_rata_target = 3.39

# Inisialisasi variabel untuk menyimpan data
data = []

# Total dari data yang telah di-generate
total_data = 0

# Generate data random dan hitung totalnya
for _ in range(jumlah_data):
    random_value = generate_random_data()
    data.append(random_value)
    total_data += random_value

# Hitung rata-ratanya
rata_rata_hasil = total_data / jumlah_data

# Koreksi rata-rata data agar sesuai dengan target
# Perhatikan bahwa data terakhir akan disesuaikan sehingga total data tetap sama.
data[-1] += rata_rata_target - rata_rata_hasil

# Cetak hasil data dan rata-ratanya
print("Data random:", data)
print("Rata-rata:", rata_rata_target)
```
